```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

tips = sns.load_dataset("tips")

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plot 1 - Total Bill vs. Tip Amount
axes[0, 0].scatter(tips['total_bill'], tips['tip'])
axes[0, 0].set_title('Total Bill vs. Tip Amount')
axes[0, 0].set_xlabel('Total Bill')
axes[0, 0].set_ylabel('Tip Amount')

# Plot 2 - Distribution of Total Bill
axes[0, 1].hist(tips['total_bill'], bins=20, color='skyblue', edgecolor='black')
axes[0, 1].set_title('Distribution of Total Bill')
axes[0, 1].set_xlabel('Total Bill')
axes[0, 1].set_ylabel('Frequency')

# Plot 3 - Boxplot of Day-wise Total Bill
sns.boxplot(x='day', y='total_bill', data=tips, ax=axes[1, 0])
axes[1, 0].set_title('Boxplot of Day-wise Total Bill')
axes[1, 0].set_xlabel('Day')
axes[1, 0].set_ylabel('Total Bill')

# Plot 4 - Violin plot of Gender-wise Tip Amount
sns.violinplot(x='sex', y='tip', data=tips, ax=axes[1, 1])
axes[1, 1].set_title('Violin plot of Gender-wise Tip Amount')
axes[1, 1].set_xlabel('Gender')
axes[1, 1].set_ylabel('Tip Amount')


plt.show()
```

```python
#GridSpec()
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib import gridspec

tips = sns.load_dataset("tips")

fig = plt.figure(figsize=(12, 10))
gs = gridspec.GridSpec(2, 2, width_ratios=[1, 1], height_ratios=[1, 1])

# Total Bill vs. Tip Amount
ax0 = plt.subplot(gs[0])
ax0.scatter(tips['total_bill'], tips['tip'])
ax0.set_title('Total Bill vs. Tip Amount')
ax0.set_xlabel('Total Bill')
ax0.set_ylabel('Tip Amount')

# Distribution of Total Bill
ax1 = plt.subplot(gs[1])
ax1.hist(tips['total_bill'], bins=20, color='skyblue', edgecolor='black')
ax1.set_title('Distribution of Total Bill')
ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Frequency')

# Boxplot of Day-wise Total Bill
ax2 = plt.subplot(gs[2])
sns.boxplot(x='day', y='total_bill', data=tips, ax=ax2)
ax2.set_title('Boxplot of Day-wise Total Bill')
ax2.set_xlabel('Day')
ax2.set_ylabel('Total Bill')

#Violin plot of Gender-wise Tip Amount
ax3 = plt.subplot(gs[3])
sns.violinplot(x='sex', y='tip', data=tips, ax=ax3)
ax3.set_title('Violin plot of Gender-wise Tip Amount')
ax3.set_xlabel('Gender')
ax3.set_ylabel('Tip Amount')

plt.tight_layout()
plt.show()
```
