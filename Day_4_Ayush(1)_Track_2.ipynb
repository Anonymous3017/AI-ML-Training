```python
# Scraping
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://realpython.github.io/fake-jobs/')
```

```python
r.content
r.status_code
```

```python
soup=BeautifulSoup(r.content,) #parser LXML, html
```

```python
soup
```

```python
soup.prettify()
```

```python
#find and find_all
job_results = soup.find(id="ResultsContainer")
```

```python
job_results
```

```python
all_jobs = job_results.find_all('div',class_='card-content')
```

```python
all_jobs[0].find('h2',class_="title is-5")
```

```python
all_jobs[0].find('h2',class_="title is-5").text
```

```python
all_jobs[0].find('h3',class_="subtitle is-6 company").text
```

```python
all_jobs[0].find('p',class_="location").text
```

```python
job_info = []
for jobs in all_jobs:
  t=jobs.find('h2',class_="title is-5").string
  c=jobs.find('h3',class_="subtitle is-6 company").string
  l=jobs.find('p',class_="location").string
  job_info.append({"Job Title":t,"Company":c,"Location":l})
```

```python
job_info
```

```python
import pandas as pd
```

```python
job_info_df = pd.DataFrame(job_info)
```

```python
job_info_df["Location"]=[i.strip() for i in job_info_df["Location"]]
```

```python
job_info_df
```

```python
python_jobs=job_results.find_all('h2',string=lambda text:"python" in text.lower())
python_job_info = []
for jobs in python_jobs:
  t=jobs.text
  c=jobs.find_next('h3',class_="subtitle is-6 company").string
  l=jobs.find_next('p',class_="location").string.split()
  a=jobs.find_next('a',class_='card-footer-item')['href']
  python_job_info.append({"Job Title":t,"Company":c,"Location":l,"Apply_link": a})

python_job_info_df = pd.DataFrame(python_job_info)

python_job_info_df
```
