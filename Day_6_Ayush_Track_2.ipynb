**Q1. Problem Statement: Univariate EDA**

Load the “cereal.csv” data into a DataFrame and perform the following tasks:

1.	Explore the DataFrame using info() and describe() functions

2.	Replace the manufacturer names with the dictionary of names given here –
{'A':'American Home Food Products', 'G':'General Mills', 'K':'Kelloggs','N':'Nabisco', 'P':'Post', 'Q':'Quaker Oats', 'R':'Ralston Purina'}

3.	Find out the top five cereal manufacturers based on customer ratings

4.	With the help of a barplot, compare the manufacturers based on customer ratings

5.	With the help of a pie chart, find out which of the manufacturers are the largest producers of cereals


6.	Find out cereal manufacturers whose products are rich in protein but contains minimal calories, with the help of a scatter plot


7.	Using a heatmap, plot the correlation and covariance between all the nutritional values present in the cereals



```python
import pandas as pd
data = pd.read_csv('/content/cereal.csv')
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# Replace the manufacturer names with the dictionary of names given here
replacement_dict = {'A':'American Home Food Products', 'G':'General Mills', 'K':'Kelloggs','N':'Nabisco', 'P':'Post', 'Q':'Quaker Oats', 'R':'Ralston Purina'}
df['mfr'] = df['mfr'].replace(replacement_dict)

df
```

```python
# Find out the top five cereal manufacturers based on customer ratings

manufacturer_ratings = df.groupby('mfr')['rating'].mean()

sorted_manufacturers = manufacturer_ratings.sort_values(ascending=False)

top_five_manufacturers = sorted_manufacturers.head(5)

top_five_manufacturers
```

```python
# With the help of a barplot, compare the manufacturers based on customer ratings
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))

plt.bar(manufacturer_ratings.index, manufacturer_ratings)

plt.xlabel('Manufacturer')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Mean Rating')
plt.title('Mean Rating of Cereal Manufacturers')
plt.show()
```

```python
# With the help of a pie chart, find out which of the manufacturers are the largest producers of cereals

manufacturer_counts = df['mfr'].value_counts()
print(manufacturer_counts)
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%')

plt.title('Cereal Production by Manufacturer')
plt.show()
```

```python
# Scatter plot for protein vs calories
plt.scatter(df['protein'], df['calories'], c='blue', alpha=0.7)

# Adding labels and title
plt.xlabel('Protein (g)')
plt.ylabel('Calories')
plt.title('Protein vs Calories for Cereals')

# # Annotating points with cereal names
# for i, txt in enumerate(df['name']):
#     plt.annotate(txt, (df['protein'][i], df['calories'][i]))

# Display the plot
plt.show()
```

```python
import seaborn as sns

# Calculate correlation and covariance matrices
correlation_matrix = df.corr()
covariance_matrix = df.cov()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[0])
axes[0].set_title('Correlation Matrix')

sns.heatmap(covariance_matrix, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[1])
axes[1].set_title('Covariance Matrix')

plt.tight_layout()
plt.show()
```

#Web scraping

Write  a  Python  program  that  can  extract  the  data  from  a  website  using  web scrapping concepts to perform the following tasks:

1.Use the request library and the link to extract the data.

2.Use BeautifulSoup to prepare the website's source code, then try to find a table on the source page.

3.After finding the table, extract data from all available columns and store it in the dataframe.Website Link:Use the below link to get the data from the table on the website.

https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population


```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population')

r.content
r.status_code
```

```python
soup=BeautifulSoup(r.content,)
```

```python
soup
```

```python
soup.prettify()
```

```python
#find and find_all
table = soup.find_all( "table" )[0]
```

```python
table
```

```python
rows = table.find_all('tr')
data = []
for row in rows[1:]:
    cols = row.find_all(['th', 'td'])
    cols = [col.text.strip() for col in cols]
    data.append(cols)
```

```python
df = pd.DataFrame(data, columns=['Rank', 'Country (or dependent territory)', 'Population', '% of world population', 'Date', 'Source', 'Unnamed'])
df.dropna(inplace=True)
df.drop(columns=['Unnamed'], inplace=True)
df.reset_index(drop=True, inplace=True)
```
