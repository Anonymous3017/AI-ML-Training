Problem statement(Probability of an event)

Consider that you have a coin and dice. Each time you roll them, you get a result, so try to solve the given questions based on coins and dice.

1.Roll the dice twice and find the probability of getting the number 6 each time.

2.Roll two dice at a time and find the probability of getting a sum of both dice greater than 9.

3.Flip a coin three times continuously and find the probability of getting three heads in a row.

4.Flip a coin anddice simultaneously and find the probability of getting head on the coin and number 5 on the dice simultaneously.

5.Card is drawn at random from the pack of 52 cards:

I.Find the probability that it is an honor card.

II.It is a face card of a diamond house.

```python
import numpy as np
import pandas as pd
```

```python
# Define a function to simulate rolling a dice
def roll_dice():
    return np.random.randint(1, 7)
```

```python
def flip_coin():
    return np.random.choice(['H', 'T'])
```

```python
# 1. Roll the dice twice and find the probability of getting the number 6 each time.
def prob_two_sixes():
    trials = 100000
    count = 0
    for _ in range(trials):
        if roll_dice() == 6 and roll_dice() == 6:
            count += 1
    return count / trials
prob_two_sixes()
```

```python
# 2. Roll two dice at a time and find the probability of getting a sum of both dice greater than 9.
def prob_sum_greater_than_9():
    trials = 100000
    count = 0
    for _ in range(trials):
        if roll_dice() + roll_dice() > 9:
            count += 1
    return count / trials

prob_sum_greater_than_9()
```

```python
# 3. Flip a coin three times continuously and find the probability of getting three heads in a row.
def prob_three_heads():
    trials = 100000
    count = 0
    for _ in range(trials):
        if flip_coin() == 'H' and flip_coin() == 'H' and flip_coin() == 'H':
            count += 1
    return count / trials
prob_three_heads()
```

```python
# 4. Flip a coin and dice simultaneously and find the probability of getting head on the coin and number 5 on the dice simultaneously.
def prob_head_and_5():
    trials = 100000
    count = 0
    for _ in range(trials):
        if flip_coin() == 'H' and roll_dice() == 5:
            count += 1
    return count / trials
prob_head_and_5()
```

```python
# 5. Card is drawn at random from the pack of 52 cards:

# I. Find the probability that it is an honor card.
total_cards = 52
honor_cards = 16
prob_honor_card = honor_cards / total_cards
print("Probability of drawing an honor card:", prob_honor_card)
```

```python
diamond_face_cards = 3
prob_diamond_face_card = diamond_face_cards / total_cards
print("Probability of drawing a face card of a diamond house:", prob_diamond_face_card)
```

```python
total_students = 100
total_boys = 23 + 17 + 10
total_girls = total_students - total_boys

maths_boys = 23
maths_girls = 15

science_boys = 17
science_girls = 23

english_boys = 10
english_girls = 12

# 1. Probability that a student loves Science subject
prob_science = (science_boys + science_girls) / total_students
print("Probability that a student loves Science subject:", prob_science)

```

```python
# 2. Probability that a student is a girl and loves Maths subject or English subject

prob_girls_maths_or_english = (maths_girls + english_girls) / total_students
print("Probability that a student is a girl and loves Maths subject or English subject:", prob_girls_maths_or_english)
```

```python
# 3. Probability that a student loves English if she is a girl, or loves Maths if he is a boy

prob_english_girls = english_girls / total_girls
prob_maths_boys = maths_boys / total_boys
prob_loves_english_or_maths = prob_english_girls + prob_maths_boys
print("Probability that a student loves English if she is a girl, or loves Maths if he is a boy:", prob_loves_english_or_maths)
```

```python
# 2. Create a DataFrame with columns for June and July rainfall greater than 500mm and whether it flooded that year

df["FLOODS"] = (df["JUN"] > 500) | (df["JUL"] > 500)
flood_prob_df = df[["YEAR", "JUN", "JUL", "FLOODS"]].copy()
flood_prob_df["JUN_GT_500"] = flood_prob_df["JUN"] > 500
flood_prob_df["JUL_GT_500"] = flood_prob_df["JUL"] > 500
flood_prob_df.head()
```

```python
# 3. Calculate the probability of flood given it rained more than 500 mm in June (P(A|B))
prob_flood_given_june_gt_500 = flood_prob_df[flood_prob_df["JUN_GT_500"]]["FLOODS"].mean()

print("\nProbability of flood given it rained more than 500 mm in June ", prob_flood_given_june_gt_500)
```

```python
# 4. Calculate the probability of rain more than 500 mm in June, given it flooded that year (P(B|A))
prob_june_gt_500_given_flood = flood_prob_df[flood_prob_df["FLOODS"]]["JUN_GT_500"].mean()
print("\nProbability of rain more than 500 mm in June, given it flooded that year :", prob_june_gt_500_given_flood)
```

```python

# 5. Probability of flood given it rained more than 500 mm in July

prob_flood_given_july_gt_500 = flood_prob_df[flood_prob_df["JUL_GT_500"]]["FLOODS"].mean()

print("\nProbability of flood given it rained more than 500 mm in July:", prob_flood_given_july_gt_500)
```

```python
# 6. Probability of rain more than 500 mm in July given it flooded that year (P(B|A))

prob_july_gt_500_given_flood = flood_prob_df[flood_prob_df["FLOODS"]]["JUL_GT_500"].mean()

print("\nProbability of rain more than 500 mm in July given it flooded that year :", prob_july_gt_500_given_flood)
```
