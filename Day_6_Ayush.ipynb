```python
df2 = pd.read_csv('/content/Books.csv')
df2
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2['Avg_Rating'] = df2.groupby("Author")["User Rating"].transform(mean)
df2
```

```python
df2["Avg_price"] = df2.groupby("Author")["Price"].transform(mean)
df2
```

```python
no_of_books_wrote = df2.groupby("Author")['Name'].count()
```

```python
most_book_wrote = no_of_books_wrote.idxmax()
most_book_wrote
```
